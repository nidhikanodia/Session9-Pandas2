```python
#Question 1: Delete unnamed columns

#import the pandas package
import pandas as pd

#Read the data in csv format
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv", sep=',')

#Print first 5 rows of data to see the header/contnets
print("Dataframe before removing unnamed columns:\n", df.head())

#Iterate through the list of columns to find the unnamed columns
for col in df.columns:
    #Check if thee current column is unnamed column
    if col.find('Unnamed') == 0:
        #If the current column is unnamed column, drop it from the original dataframe using inplace = True
        df.drop([col], axis = 1, inplace = True)
        
#Print the first 5 rows of data to see if the unnamed columns have been removed
print("Dataframe after removing unnamed columns:\n", df.head())
```

```python
#Question 2: Show the distribution of male and female

#Group by Gender column and sum the values in Count column to get the distribution of male and female
print("The distribution of Male and Female is:\n", df.groupby(['Gender']).sum()['Count'])
```

```python
#Question 3: Show the top 5 most preferred names

#Group by Name column and sume the values in Count column to get the total count for each name. Also, apply sorting on the Count column and reset the index
namedf = df.groupby(['Name']).sum().sort_values(by=['Count'],ascending=False).reset_index()

#Display the top 5 most preferred names
print("The top 5 most preferred names are:\n", namedf['Name'].head())
```

```python
#Question 4: The median name occurence in the dataset

#Use the mode function on name column to find the median name
name = df.loc[:,"Name"].mode()

#Get the value of median name as str
n = name.iloc[0]

#Filter the dataframe by median name and count the occurrence of median name
df1 = df[df["Name"] == n].groupby("Name").count().reset_index()

#Display the median name occurrence in the dataset
print("The median name occurrence in dataset is:\n", df1[['Name','Count']])

```

```python
#Question 5: Distribution of male and female born count by states

#Group the data by State and Gender column and sum the values. Also, reset the index
df1 = df.groupby(['State','Gender']).sum().reset_index()

#Display the distribuion of male and female born count by states
print("Distribution of male and femail born count by State:\n", df1[['State', 'Gender','Count']])
```
